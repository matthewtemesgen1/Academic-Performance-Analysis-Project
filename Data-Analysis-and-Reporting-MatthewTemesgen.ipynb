Introduction/Motivation

My project aims to analyze and predict student academic performance on exams in higher education institutions by identifying the key factors that contribute to academic success or failure.

Research Questions:
1. What are the primary factors that predict student academic performance in higher education institutions?
2. How do demographic factors such as age, gender, and socioeconomic status impact academic performance?
3. Can early academic performance indicators predict final grades or the likelihood of course completion?

Data Collection

The data for this project was collected from the "Student Performance" dataset available on Kaggle. The dataset includes 1,000 records of students' academic performance (math, reading, and writing scores) and demographic attributes.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "final_data.csv"
data = pd.read_csv(file_path)
```

```python
print("### Data Description")
print(f"Number of Samples: {data.shape[0]}")
print("Columns/Variables:")
print(data.columns.tolist())
```

```python
print("\nBasic Statistics:")
print(data[['math_score', 'reading_score', 'writing_score']].describe())
```

Data Cleaning and Wrangling

I cleaned the dataset following these steps:
1. Standardization of categorical variables by converting text to lowercase.
2. Renaming columns for consistency and readability.
3. Validation of numerical data to ensure all scores were within the expected range (0â€“100).
4. Removal of duplicates (if any) and handling missing data (though none was present).

Data Analysis

To answer the research questions, exploratory data analysis was performed, including visualizations and statistical analysis.

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['math_score'], kde=True, color="blue", label="Math")
sns.histplot(data['reading_score'], kde=True, color="green", label="Reading")
sns.histplot(data['writing_score'], kde=True, color="red", label="Writing")
plt.title("Distribution of Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
gender_means = data.groupby('gender')[['math_score', 'reading_score', 'writing_score']].mean()
gender_means.plot(kind='bar', figsize=(10, 6))
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
correlation_matrix = data[['math_score', 'reading_score', 'writing_score']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Between Scores")
plt.show()
```

Final Observations

1. Math and reading scores were strongly correlated with writing scores, as seen in the heatmap.
2. Gender appeared to influence average scores, with differences in performance in certain subjects.
3. Parental education showed significant variation in students' performance, particularly in math.

The results suggest that demographic factors and early performance indicators can help predict academic success.
